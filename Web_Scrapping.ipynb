```python
#URL of the particular website (here it is OTC drugs website)
URL = "https://www.drugs.com/otc"

# To scrap the information present in multiple pages, we look into the URL of the pages and try to find some
# sort of similarity in their URL.
# In the OTC drug scraping problem, there is a similarity in the URL structure of different pages such as:
# "https://www.drugs.com/otc-a1.html" , "https://www.drugs.com/otc-a2.html" and so on....
# Thus, here we iterate to all of the pages simulataneously using their similarity in the URL.

for i in range(ord('a'), ord('z') + 1):
  for j in range(1, 8):
    page = requests.get(URL + "-" + chr(i) + str(j) + ".html") 
    soup = BeautifulSoup(page.content, "html.parser")
    results = soup.find("div", class_="contentBox") 

    k = results.find("h1")
    if (k.text == 'Page Not Found'):
      break
   
    elements = results.find_all("ul", class_="ddc-list-column-2")
    for element in elements:
      print(element.text)


for i in range(0, 10):
  page = requests.get(URL + "-" + str(i) + str(1) + ".html") 
  soup = BeautifulSoup(page.content, "html.parser")
  results = soup.find("div", class_="contentBox") 

  k = results.find("h1")
  if (k.text == 'Page Not Found'):
    break
  
  elements = results.find_all("ul", class_="ddc-list-column-2")
  for element in elements:
    print(element.text)
```
